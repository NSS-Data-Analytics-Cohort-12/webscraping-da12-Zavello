```python
sr_python_url = 'https://realpython.github.io/fake-jobs/jobs/senior-python-developer-0.html'

response_3 = requests.get(sr_python_url)
```

```python
sr_python = BS(response_3.text)
sr_python
```

```python
sr_python_describe = sr_python.find('div', attrs={'class' : 'content'}).find('p').text
sr_python_describe
```

```python
def job_description(url_list):
    response = requests.get(url_list)
    soups = BS(response.text)
    job_description = soups.find('div', attrs={'class' : 'content'}).find('p').text
    return job_description
   
```

```python
urls_series = pd.Series(url_list)
urls_series.info()
```

```python
jd = urls_series.apply(job_description)
jd
```

```python
job_df['Job Description'] = jd
```
